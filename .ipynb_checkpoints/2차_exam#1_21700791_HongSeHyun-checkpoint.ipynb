# Data Preprocessing

### b) (10pt) ‘CustomerID’의 값이 비어 있는 행을 제거하고, ‘Quantity’와 ‘UnitPrice’가 모두 양수인 행 만 남기기.

(Hint: ‘Quantity’가 양수인 행을 남긴 다음, ‘UnitPrice’가 양수인 행을 남기면 됨.)

```python
df.dropna(subset=['CustomerID'], axis='index', how='any', inplace=True)
df
```

```python
filt_Quantity = (df['Quantity']>0)
df = df[filt_Quantity]
df
```

```python
filt_UnitPrice = (df['UnitPrice']>0)
df = df[filt_UnitPrice]
df
```

### (c) (10pt) ‘UnitPrice’ 와 ‘Quantity’ 값을 곱하여 ‘TotalExpense’라는 새로운 열을 생성

```python
df['TotalExpense'] = np.nan
df['TotalExpense'] = df['Quantity']*df['UnitPrice']
df
```

# Pivoting Features

### (d) (10pt) ‘StockCode’ (상품 아이디) 별로, 고객들의 가장 많은 소비를 유도한 상품 아이디를 내림차순으로 출력

```python
StockCode_Group = df.groupby('StockCode').sum().sort_values('TotalExpense', ascending=False)
StockCode_Group
```

# Visualizing Data

### (e) (5pt) 고객들의 지출 총합을 전체 매출 변수 (‘total_revenue’) 에 저장하기.

```python
total_revenue = df['TotalExpense'].sum()
total_revenue
```

### (f) (15pt) 국가별 매출을 구하여 오름차순으로 정렬, 각 국가가 전체 매출의 몇 %를 차지하는지 구하기 (+시각화)

```python
Country_Group = df.groupby('Country').sum().sort_values('TotalExpense', ascending=True)
Country_Group = Country_Group.loc[ : , 'TotalExpense']

Country_Group_Pct = (Country_Group/total_revenue)*100
Country_Group_Pct.astype(str) + '%'
```

```python
Country_Group_Pct_df = pd.DataFrame([Country_Group_Pct]).T
Country_Group_Pct_df.columns = ['Revenue(%)']
Country_Group_Pct_df
```

```python
plt.figure(figsize=(15,7))
plt.title('Revenue by Country')
plt.bar(Country_Group_Pct_df.index, Country_Group_Pct_df[Country_Group_Pct_df.columns[0]])

plt.xlabel(Country_Group_Pct_df.index.name)
plt.ylabel(Country_Group_Pct_df.columns[0])

plt.xticks(rotation=45, fontsize =8)
plt.show()
```

### (g) (5pt) 자신이 불러온 데이터셋의 이름이 ‘data’ 라고 가정했을 때, 다음의 코드를 추가하여 실행한 후, InvoiceDate’ 의 datatype 이 ‘object’에서 무엇으로 바뀌었는지 확인하여라.

```python
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], infer_datetime_format=True)
```

```python
df.info()
```

### (h) (15pt) ‘연도-월’ 별 매출을 구하고, 이것을 시각화 하기.

```python
df['InvoiceDate']
```

```python
df['year'] = df['InvoiceDate'].dt.strftime('%Y')
df['month'] = df['InvoiceDate'].dt.strftime('%m')

df['Year-Month'] = df['year']+'-'+df['month']
df
```

```python
Year_Month_Revenue = df.groupby('Year-Month').sum()
Year_Month_Revenue = Year_Month_Revenue.loc[ : , 'TotalExpense']
Year_Month_Revenue
```

```python
Year_Month_Revenue_df = pd.DataFrame([Year_Month_Revenue]).T
Year_Month_Revenue_df.columns = ['Revenue']
Year_Month_Revenue_df
```

```python
plt.figure(figsize=(15,7))
plt.title('Revenue by Month')
plt.bar(Year_Month_Revenue_df.index, Year_Month_Revenue_df[Year_Month_Revenue_df.columns[0]])

plt.xlabel(Year_Month_Revenue_df.index.name)
plt.ylabel(Year_Month_Revenue_df.columns[0])

plt.xticks(rotation=45, fontsize =8)
plt.show()
```

### (i) (20pt) 문제 (f )와 (h)에서 그래프로 시각화하는 작업을 공통되게 할 수 있는 함수를 정의하기.

(함수 parameter 는 마음대로 지정할 수 있음. e.g. x, y, xlabel, ylabel, title 등)

```python
# Case (1)   DataFrame으로 불러오기
def plotting_bar(dataframe, title, title_fontsize, extra_fontsize, figure_size_X, figure_size_Y):
    plt.figure(figsize=(figure_size_X,figure_size_Y))
    
    title = str(title)
    plt.title(title, fontsize=title_fontsize)
    
    plt.bar(dataframe.index, dataframe[dataframe.columns[0]])

    plt.xlabel(dataframe.index.name, fontsize=extra_fontsize)
    plt.ylabel(dataframe.columns[0], fontsize=extra_fontsize)
    plt.xticks(rotation=45, fontsize=extra_fontsize)
    
    plt.show()
```

```python
plotting_bar(Country_Group_Pct_df, 'Revenue by Count', 25, 10, 20, 10)
```

```python
plotting_bar(Year_Month_Revenue_df, 'Revenue by Month', 25, 15, 20, 10)
```

```python
# Case (2)   문제 예시와 같이 각 변수로 접근하기
def plotting_bar(x_data, y_data, xlabel, ylabel, title, title_fontsize, extra_fontsize, figure_size_X, figure_size_Y):
    plt.figure(figsize=(figure_size_X,figure_size_Y))
    
    title = str(title)
    plt.title(title, fontsize=title_fontsize)
    
    plt.bar(x_data, y_data)

    plt.xlabel(xlabel, fontsize=extra_fontsize)
    plt.ylabel(ylabel, fontsize=extra_fontsize)
    plt.xticks(rotation=45, fontsize=extra_fontsize)
    
    plt.show()
```

```python
plotting_bar(Country_Group_Pct_df.index, Country_Group_Pct_df[Country_Group_Pct_df.columns[0]],
             Country_Group_Pct_df.index.name, Country_Group_Pct_df.columns[0], 'Revenue by Count', 25, 10, 20, 10)
```

```python
plotting_bar(Year_Month_Revenue_df.index, Year_Month_Revenue_df[Year_Month_Revenue_df.columns[0]],
             Year_Month_Revenue_df.index.name, Year_Month_Revenue_df.columns[0], 'Revenue by Month', 25, 10, 20, 10)
```
