```python
def func_age_category():
    df['age_category'] = ((df['Age']//10)*10).astype(int)

func_age_category()
df = df.reindex(columns=['Survived','Pclass','Sex','Age','age_category','SibSp','Parch','Fare','Cabin','Embarked'])

df['age_category'] = df['age_category'].astype(str) + '대'
df
```

```python
# 주어진 그래프 처럼 Plotting 해보기
Sur_O_data = df[df['Survived']==1].loc[:,'Age']
Sur_X_data = df[df['Survived']==0].loc[:,'Age']

plt.hist([Sur_X_data, Sur_O_data], label=['Survived: X', 'Survived: O'])

plt.title('Age - Survival count')
plt.xlabel('age')
plt.ylabel('count')
plt.xticks( [i for i in range(0,81,10)] )   # 각 인덱스에 각각의 이름을 배치해주는 것
plt.legend()
plt.show()
```

```python
# step(1)  연령대별 인원수 카운팅
Num_people = df.groupby('age_category').count()
Num_people = Num_people.loc[ : , 'Survived']

# step(2) 연령대별 생존자 카운팅
O_people = df.groupby('age_category').sum()
O_people = O_people.loc[ : , 'Survived']

# step(3) 연령대별 사망자 카운팅
X_people = Num_people-O_people

# step(4) Plotting을 위한 DataFrame생성
df_survive = pd.DataFrame([O_people, X_people])
df_survive.index=['생존자','사망자']

df_survive = df_survive.T
df_survive



# method(2) - DataFrame을 생성하는 또 다른 방법(2)
# data_1 = {
#     '생존자' : [O_people[0],O_people[1],O_people[2],O_people[3],O_people[4],O_people[5],O_people[6],O_people[7],O_people[8]],
#     '사망자' : [X_people[0],X_people[1],X_people[2],X_people[3],X_people[4],X_people[5],X_people[6],X_people[7],X_people[8]]
# }

# df_survive = pd.DataFrame(data_1, index = [i for i in range(0,81,10)])
# df_survive
```

```python
# step(5) 히스토그램 생성
N = df_survive.shape[0]           # 연령대별 평균값 테이블의 행의 수를 가져옴 = 9개
index = np.arange(N)*10           # df_survive dataframe의 index값이 위와 같이 숫자가 아니라면,
                                  # 가중치에 의한 막대 그래프의 횡 이동이 어려우므로 이 과정이 필요하다.

width = 3.5                       # 폭
w = 3.5                           # 가중치

plt.figure(figsize=(8,4))
plt.title('Age - Survival count')
plt.bar(index+w,       df_survive['사망자'], width = width, label='Survived: X')
plt.bar(index+w+width, df_survive['생존자'], width = width, label='Survived: O')

plt.xlabel('age')
plt.ylabel('count')
plt.xlim([0,90])
plt.xticks([i for i in range(0,91,10)] )   # 각 인덱스에 각각의 이름을 배치해주는 것
plt.yticks([i for i in range(0,141,20)] )
plt.legend()
plt.grid()
plt.show()

# method(2) - dataframe을 바로 plotting하는 방법
# df_survive.plot(kind='bar', stacked=False)
```

```python
# 주어진 그래프 처럼 Plotting 해보기
men = df[df['Sex'] == 'male']
women = df[df['Sex'] == 'female']

Sur_O_Mdata = men[men['Survived']==1].loc[:,'Age']
Sur_O_Wdata = women[women['Survived']==1].loc[:,'Age']
Sur_X_Mdata = men[men['Survived']==0].loc[:,'Age']
Sur_X_Wdata = women[women['Survived']==0].loc[:,'Age']

fig, axs = plt.subplots(1, 2, figsize=(20, 7))

axs[0].hist([Sur_X_Mdata, Sur_O_Mdata], label=['Survived: X', 'Survived: O'])
axs[1].hist([Sur_X_Wdata, Sur_O_Wdata], label=['Survived: X', 'Survived: O'])
axs[0].legend()                                        # 범례 표시
axs[1].legend()                                        # 범례 표시
plt.show()
```

```python
# step(1) 남녀 - 연령대별 사람 수 구하기
men_num = men.groupby(['Sex','age_category']).count().loc[:, 'Survived']
women_num = women.groupby(['Sex','age_category']).count().loc[:, 'Survived']

people_num = (men_num, women_num)
people_num
```

```python
# step(2) 남녀 - 연령대별 생존자 수 구하기
O_men = men.groupby(['Sex','age_category']).sum().loc[:, 'Survived']
O_women = women.groupby(['Sex','age_category']).sum().loc[:, 'Survived']

Sur_O = (O_men, O_women)
Sur_O
```

```python
# step(3) 남녀 - 연령대별 사망자 수 구하기
X_men = people_num[0] - Sur_O[0]
X_women = people_num[1] - Sur_O[1]

Sur_X = (X_men, X_women)
Sur_X
```

```python
data_2 = {
    '남자생존자' : [Sur_O[0][0],Sur_O[0][1],Sur_O[0][2],Sur_O[0][3],Sur_O[0][4],Sur_O[0][5],Sur_O[0][6],Sur_O[0][7],Sur_O[0][8]],
    '남자사망자' : [Sur_X[0][0],Sur_X[0][1],Sur_X[0][2],Sur_X[0][3],Sur_X[0][4],Sur_X[0][5],Sur_X[0][6],Sur_X[0][7],Sur_X[0][8]]
}

df_men = pd.DataFrame(data_2, index = [i for i in range(0,81,10)])
df_men.index.name = '연령대'
df_men
```

```python
data_3 = {
    '여자생존자' : [Sur_O[1][0],Sur_O[1][1],Sur_O[1][2],Sur_O[1][3],Sur_O[1][4],Sur_O[1][5],Sur_O[1][6]],     
    '여자사망자' : [Sur_X[1][0],Sur_X[1][1],Sur_X[1][2],Sur_X[1][3],Sur_X[1][4],Sur_X[1][5],Sur_X[1][6]]
}

df_women= pd.DataFrame(data_3, index = [i for i in range(0,61,10)])
df_women.index.name = '연령대'
df_women
```

```python
# step(5) 히스토그램 생성
fig, axs = plt.subplots(1, 2, figsize=(20, 7))

width = 3.5                         # 폭
w = 3.5                             # 가중치

# 첫번째 그래프 넣기
axs[0].bar(df_men.index+w,       df_men['남자사망자'], width=width, label='Survived: X')   # 데이터 설정
axs[0].bar(df_men.index+w+width, df_men['남자생존자'], width=width, label='Survived: O')   # 데이터 설정
axs[0].set_title('Male')                                                             # 제목 설정

# 두 번째 그래프 넣기
axs[1].bar(df_women.index+w,       df_women['여자사망자'], width=width, label='Survived: X')   # 데이터 설정
axs[1].bar(df_women.index+w+width, df_women['여자생존자'], width=width, label='Survived: O')   # 데이터 설정
axs[1].set_title('Female')                                                               # 제목 설정

for i in range(0,2):
    axs[i].set(xlabel='age', ylabel='count')
    axs[i].grid(linestyle='-', linewidth=1)
    axs[i].set_xlim([0,90])
    axs[i].set_xticks([i for i in range(0,91,10)])
    axs[i].set_yticks([i for i in range(0,121,20)])
    axs[i].set_ylim(0,130)  
    axs[i].legend()  

plt.show()
```

```python
# step(1) Titanic 호에 동승한 자매 또는 배우자의 수의 유무에 따른 생존과 사망 데이터 가져오기
O_SibSp = df[df['SibSp'] > 0]['Survived'].value_counts()
X_SibSp = df[df['SibSp'] == 0]['Survived'].value_counts()

# step(2) Titanic 호에 동승한 부모 또는 자식의 수의 유무에 따른 생존과 사망 데이터 가져오기
O_Parch = df[df['Parch'] > 0]['Survived'].value_counts()
X_Parch = df[df['Parch'] == 0]['Survived'].value_counts()

SibSp_Parch_Survive_Relation = pd.DataFrame([O_SibSp, X_SibSp, O_Parch, X_Parch])
SibSp_Parch_Survive_Relation.index=['동승한 자매 또는 배우자가 있는 경우', '동승한 자매 또는 배우자가 없는 경우', '동승한 부모 또는 자식이 있는 경우', '동승한 부모 또는 자식이 없는 경우']
SibSp_Parch_Survive_Relation.columns=['사망','생존']
SibSp_Parch_Survive_Relation
```

```python
SibSp_Parch_Survive_Relation.plot(kind='bar')
```

```python
# 생존률 비교
rate = (SibSp_Parch_Survive_Relation['생존']/(SibSp_Parch_Survive_Relation['사망']+SibSp_Parch_Survive_Relation['생존']))*100
rate = rate.astype(str)+'%'
rate
```

```python
fig, axs_1 = plt.subplots(figsize=(10,7)
                         )
axs_1.plot(df2.index, df2['출생아 수'], color='#ff812d')

axs_2 = axs_1.twinx()                     # x축을 공유하는 쌍둥이 axis
axs_2.plot(df2.index, df2['합계 출산율'], color='#ffd100')

plt.show()
```

```python
fig, axs_1 = plt.subplots(figsize=(13,5))
fig.suptitle('출생아 수 및 합계출산율')

axs_1.bar(df2.index, df2['출생아 수'], color='#ff812d')
axs_1.set_ylabel('출생아 수(천 명)')
axs_1.set_ylim(250,700)
axs_1.set_yticks([300,400,500,600])

for idx, val in enumerate(df2['출생아 수']):
    axs_1.text(idx, val+12, val, ha='center')


axs_2 = axs_1.twinx()                     # x축을 공유하는 쌍둥이 axis
axs_2.plot(df2.index,df2['합계 출산율'],color='#ffd100',marker='o',markersize=10,markeredgecolor='w',markeredgewidth=3,linewidth=3)
axs_2.set_ylabel('합계 출산율(가임여성 1명당 명)')
axs_2.set_ylim(0,1.5)
axs_2.set_yticks([0,1])

for idx, val in enumerate(df2['합계 출산율']):
    axs_2.text(idx, val+0.08, val, ha='center')

plt.show()
```
