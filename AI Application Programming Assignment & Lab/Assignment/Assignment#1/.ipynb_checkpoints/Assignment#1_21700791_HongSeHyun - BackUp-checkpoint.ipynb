### (d) 나이를 구간별로 나눌 수 있는 함수를 정의하고, 나이 구간 별 평균 생존율을 구한 뒤, 오름차순으로 정렬
* 주의: 나이를 구간별로 나눈 ‘Age’열을 원본의 ‘Age’ 열에 덮어씌우고, (e)와 (f)의 그래프를 그리도록 한다.


       └ 이 부분의 경우, 나이를 구간별로 나눈 값을 저장할 열을 새롭게 정의하고
          (e)와 (f)의 그래프에 적용해보겠다.
     
     이렇게 하는 이유는, 'Age'열을 살려두고, 연령대별로 구분해주는 'age_category'열을 만듦으로써
     연령대가 정확하게 계산이 되었는지 확인하기 위함이다.
     
     따라서, 과제에서 주어진 주의 조건과는 약간 다르지만,
     이 방식으로 진행하도록 하겠다.
     
    

##### Bonus) 이 과정을 통해 알 수 있는 사실
    [1] 연령대별 생존률만 놓고 보았을 때,
        영유아의 경우 생존률이 압도적으로 높다는 것을 확인할 수 있다.
        
    [2] 70대와 80대를 살펴보면 생존률이 각각 0%, 100%로 나타났는데,
        이 경우, 데이터가 너무 적어서 확률이 비정상적으로 낮거나 크게 나오지 않았나 의심을 할 필요가 있다.
        
    [3] 이 외의 다른 연령대에 대해서는, 30~40% 대에서 비교적 고른 생존률을 보였다.
        따라서, 10~60대에 대해서는 '연령대별 생존률'에 대한 데이터로는 정확한 비교와 분석이 어렵지 않나 생각한다. 

```python
def func_age_category():
    df['age_category'] = ((df['Age']//10)*10).astype(int)

func_age_category()
df = df.reindex(columns=['Survived','Pclass','Sex','Age','age_category','SibSp','Parch','Fare','Cabin','Embarked'])

df['age_category'] = df['age_category'].astype(str) + '대'
df
```

```python
age_group = df.groupby('age_category').mean().sort_values('Survived')
age_group.loc[ : , 'Survived'] * 100
```

### (e) 사망자와 생존자 각각을 나이별 분포로 나타낸 히스토그램 그리기.

```python
# 주어진 그래프 처럼 Plotting 해보기
Sur_O_data = df[df['Survived']==1].loc[:,'Age']
Sur_X_data = df[df['Survived']==0].loc[:,'Age']

plt.hist([Sur_X_data, Sur_O_data], label=['Survived: X', 'Survived: O'])

plt.title('Age - Survival count')
plt.xlabel('age')
plt.ylabel('count')
plt.xticks( [i for i in range(0,81,10)] )   # 각 인덱스에 각각의 이름을 배치해주는 것
plt.legend()
plt.show()
```

#### 과제 내용이 바뀌기 전인 위 방식의 경우, 연령대 구분이 명확히 되지 않기 때문에 좋은 그래프라 하기 어렵다.
#### 따라서, 조금 더 Develop하여 다음과 같은 순서 하에 Plotting을 진행하였다.


* step(1)  연령대별 인원수 카운팅
* step(2) 연령대별 생존자 카운팅
* step(3) 연령대별 사망자 카운팅
* step(4) Plotting을 위한 DataFrame생성

##### Bonus) 이 과정을 통해 알 수 있는 사실
    [1] 타이타닉호 탑승객 중에서 20~30대가 차지하는 비중이 큼을 확인할 수 있다.
    
    [2] 70~80대의 탑승객은 많지 않았음을 확인할 수 있다.
  
    [3] 타이타닉호 탑승객 중에서 20~30대가 차지하는 비중이 커서 그런지
        20~30대 생존자가 많음을 확인할 수 있다.
        
    [4] 마찬가지로,
        타이타닉호 탑승객 중에서 20~30대가 차지하는 비중이 커서 그런지
        20~30대 사망자가 많음을 확인할 수 있다.

```python
# step(1)  연령대별 인원수 카운팅
Num_people = df.groupby('age_category').count()
Num_people = Num_people.loc[ : , 'Survived']

# step(2) 연령대별 생존자 카운팅
O_people = df.groupby('age_category').sum()
O_people = O_people.loc[ : , 'Survived']

# step(3) 연령대별 사망자 카운팅
X_people = Num_people-O_people

# step(4) Plotting을 위한 DataFrame생성
df_survive = pd.DataFrame([O_people, X_people])
df_survive.index=['생존자','사망자']

df_survive = df_survive.T
df_survive



# method(2) - DataFrame을 생성하는 또 다른 방법(2)
# data_1 = {
#     '생존자' : [O_people[0],O_people[1],O_people[2],O_people[3],O_people[4],O_people[5],O_people[6],O_people[7],O_people[8]],
#     '사망자' : [X_people[0],X_people[1],X_people[2],X_people[3],X_people[4],X_people[5],X_people[6],X_people[7],X_people[8]]
# }

# df_survive = pd.DataFrame(data_1, index = [i for i in range(0,81,10)])
# df_survive
```

* step(5) 히스토그램 생성

##### Bonus) 이 과정을 통해 알 수 있는 사실
        앞에서 살펴본 '연령대별 생존률'만 놓고보았을 때는,
        '영유아의 생존률이 압도적으로 높다'
        '70~80대 생존률의 경우, 생존률이 각각 0%와 100%로 나타났는데, 데이터가 너무 적지는 않았나 의심이 된다'
        '10~60대에 대해서는 비교적 고른 생존률을 보였기에, 정확한 분석이 어렵다' 라는 결론을 내릴 수 있었다.
        
        이번에는, '연령대에 따른 사망자 수와 생존자 수를 나타낸 그래프'를 살펴보자면
        다음과 같은 결론을 내릴 수 있다.
        
        
    [1] 영유아:  영유아에 해당하는 탑승객이 60명 정도 되므로, 영유아에 대한 데이터는 적지 않다고 볼 수 있으며
                 그래프를 살펴보면, 사망자보다 생존자가 많은 것으로 보아, 확실히 영유아를 우선적으로 구출했음을 확인할 수 있다.
                 
    [2] 70~80대: 70대에 대한 데이터는 6건, 80대에 대한 데이터는 1건이므로, 이를 바탕으로 일반화하기에는 무리가 있다고 볼 수 있다.
                 한 가지 예를 들자면,
                     '80대의 생존률은 100%다. 따라서 고령일수록 우선적으로 구출될 수 있었다.'라고 일반화할 수 없다는 것이다.
                     똑같이 고령층에 속하는 60대와 70대의 데이터를 살펴보면 오히려 생존자 대비 사망자가 월등히 많다.
                     
    [3] 10~60대: 10~60대 모두 생존자보다 사망자가 많았다.
                 20대의 경우, 사망자를 나타내는 파란색 그래프와 생존자를 나타내는 황색 그래프 사이의 간격이 크다.
                        이는, 생존자보다 사망자가 훨씬 많다는 것을 의미한다.
                 30대의 경우, 사망자를 나타내는 파란색 그래프와 생존자를 나타내는 황색 그래프 사이의 간격이 적다.
                        이는, 생존자보다 사망자가 많지만 20대와 40대 보다는 적다는 것을 의미한다.
                 40대의 경우, 사망자를 나타내는 파란색 그래프와 생존자를 나타내는 황색 그래프 사이의 간격이 크다.
                        이는, 생존자보다 사망자가 훨씬 많다는 것을 의미한다.
                 50대의 경우, 사망자를 나타내는 파란색 그래프와 생존자를 나타내는 황색 그래프 사이의 간격이 적다.
                        이는, 생존자보다 사망자가 많지만 20대와 40대 보다는 적다는 것을 의미한다.
                        
                 이를 바탕으로 결론을 내리자면, '10~60대에서는 생존과 관련하여 연령대가 미치는 영향은 없다.'는 것이다.
                 만약 젊을 수록 우선적으로 구출했다면 연령대가 증가할수록 생존률이 떨어져야 할 것이다.
                 반대로, 늙을 수록 우선적으로 구출했다면 연령대가 증가할수록 생존률이 높아져야 할 것이다.
                 하지만, 10~60대에서는 이러한 형태를 가지지 않았다.

    ★ 결론: '연령대별 생존률'과 '연령대에 따른 사망자 수와 생존자 수를 나타낸 그래프'를 통해서는

### [2] 동승자가 있는 사람의 생존율과 동승자가 없는 사람의 생존율 비교

##### Bonus) 이 과정을 통해 알 수 있는 사실
    [1] 동승한 자매 또는 배우자가 있는 경우,
        사망자를 나타내는 파란색 그래프와 생존자를 나타내는 황색 그래프 사이의 간격이 적다.
        
        이에 비해,
        동승한 자매 또는 배우자가 없는 경우,
        사망자를 나타내는 파란색 그래프와 생존자를 나타내는 황색 그래프 사이의 간격이 크다.
        
        수치로 살펴보면,
        동승한 자매 또는 배우자가 있는 경우, 생존률은 47.33%를 기록했지만
        동승한 자매 또는 배우자가 없는 경우, 생존률은 37.15%를 기록했다.
        
        이를 통해,
        동승한 자매 또는 배우자가 있는 경우 비교적 구출에 있어서 우선순위 였고, 이에 따라 생존률이 높게 나왔을 것이라 생각된다.
     
     
    [2] 동승한 부모 또는 자식이 있는 경우,
        사망자를 나타내는 파란색 그래프와 생존자를 나타내는 황색 그래프 사이의 간격이 적다.
        
        이에 비해,
        동승한 부모 또는 자식이 없는 경우,
        사망자를 나타내는 파란색 그래프와 생존자를 나타내는 황색 그래프 사이의 간격이 크다.
        
        수치로 살펴보면,
        동승한 부모 또는 자식이 있는 경우, 생존률은 53.89%를 기록했지만
        동승한 부모 또는 자식이 없는 경우, 생존률은 35.7%를 기록했다.
        
        이를 통해,
        동승한 부모 또는 자식이 있는 경우 비교적 구출에 있어서 우선순위 였고, 이에 따라 생존률이 높게 나왔을 것이라 생각된다.
  
  
    즉, 동승자의 유무 역시 생존 여부에 있어서 영향을 미친다고 볼 수 있다.

```python
# step(1) Titanic 호에 동승한 자매 또는 배우자의 수의 유무에 따른 생존과 사망 데이터 가져오기
O_SibSp = df[df['SibSp'] > 0]['Survived'].value_counts()
X_SibSp = df[df['SibSp'] == 0]['Survived'].value_counts()

# step(2) Titanic 호에 동승한 부모 또는 자식의 수의 유무에 따른 생존과 사망 데이터 가져오기
O_Parch = df[df['Parch'] > 0]['Survived'].value_counts()
X_Parch = df[df['Parch'] == 0]['Survived'].value_counts()

SibSp_Parch_Survive_Relation = pd.DataFrame([O_SibSp, X_SibSp, O_Parch, X_Parch])
SibSp_Parch_Survive_Relation.index=['동승한 자매 또는 배우자가 있는 경우', '동승한 자매 또는 배우자가 없는 경우', '동승한 부모 또는 자식이 있는 경우', '동승한 부모 또는 자식이 없는 경우']
SibSp_Parch_Survive_Relation.columns=['사망','생존']
SibSp_Parch_Survive_Relation
```

```python
SibSp_Parch_Survive_Relation.plot(kind='bar')
```

```python
# 생존률 비교
rate = (SibSp_Parch_Survive_Relation['생존']/(SibSp_Parch_Survive_Relation['사망']+SibSp_Parch_Survive_Relation['생존']))*100
rate = rate.astype(str)+'%'
rate
```
